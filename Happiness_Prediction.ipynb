```python
# YOUR CODE HERE
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

```python
# YOUR CODE HERE

numeric_feats = df.select_dtypes(include=['float64', 'int64']).columns
imputer = SimpleImputer(strategy='mean')
df[numeric_feats] = imputer.fit_transform(df[numeric_feats])
```

```python
features = ['Log GDP per capita','Social support','Healthy life expectancy at birth',
'Freedom to make life choices','Generosity', 'Democratic Quality'
]
y = df['Life Ladder']
X = df[features]
```

```python
# Feature Scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)
y_pred_train_lr = model.predict(X_train)
y_pred_test_lr = model.predict(X_test)
```

```python
# Random Forest Regressor Model
model2 = RandomForestRegressor(random_state=42)
model2.fit(X_train, y_train)
y_pred_train_rf = model2.predict(X_train)
y_pred_test_rf = model.predict(X_test)
```

```python
# Evaluating the models
def evaluate_model(y_train, y_pred_train, y_test, y_pred_test, model_name):
    print(f"Evaluation for {model_name}")
    print("Training set performance")
    print(f"Mean Squared Error: {mean_squared_error(y_train, y_pred_train)}")
    print(f"R^2 Score: {r2_score(y_train, y_pred_train)}")
    print("\nTesting set performance:")
    print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred_test)}")
    print(f"R^2 Score: {r2_score(y_test, y_pred_test)}")
    print("\n")
```

```python
# Evaluate Linear Regression Model
evaluate_model(y_train, y_pred_train_lr, y_test, y_pred_test_lr, "Linear model")
```

```python
# Evaluate Random Forest Regressor Model
evaluate_model(y_train, y_pred_train_rf, y_test, y_pred_test_rf, "Random Forest Model")
```

```python
#The Linear Regression model shows a decent performance with an R score of 0.733 on the training set and 0.723 on the testing set. 
#This indicates that the model explains about 73% of the variance in the target variable on both the training and testing sets, suggesting a good fit without overfitting.
```

```python
#The Random Forest model exhibits a very high R score of 0.979 on the training set, which indicates almost perfect fitting on the training data. 
#However, the R score drops to 0.723 on the testing set, which is the same as the Linear Regression model's performance on the testing set.


#The MSE on the training set is significantly lower for the Random Forest model (0.027) compared to the Linear Regression model (0.333),
#Indicating that the Random Forest is overfitting the training data.


#The similar MSE (0.358) and R score (0.723) on the testing set for both models indicate that the Random Forest model does not generalize any better than the Linear Regression model, 
#despite its complexity and overfitting on the training data.
```

```python
# Hyperparameter Tuning for Random Forest Model
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(estimator=model2, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
best_rf_model = grid_search.best_estimator_

y_pred_train_best_rf = best_rf_model.predict(X_train)
y_pred_test_best_rf = best_rf_model.predict(X_test)
```

```python
evaluate_model(y_train, y_pred_train_best_rf, y_test, y_pred_test_best_rf, "Tuned Random Forest Model")
```

```python
# The tuned Random Forest model has improved the generalization performance significantly while maintaining high training performance. 
# This balance suggests that the model is now better at predicting new data while still fitting the training data well.

#Eventhough the training performance is still a bit high, the significant improvement in test performance suggests that overfitting has been reduced compared to the initial untuned model.
```
